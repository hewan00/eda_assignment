```python
# --- Q2 Airbnb Data ---
df_airbnb = pd.read_csv('./eda_assignment/data/airbnb_hw.csv', low_memory=False)

# background info
print("Shape of Airbnb DF:", df_airbnb.shape)
print("Cols:", df_airbnb.columns.tolist())
df_airbnb.head()
```

```python
pd.crosstab(df_airbnb['Room Type'], df_airbnb['Property Type'])
```

```python
# Convert price to numeric (some commas/strings)
df_airbnb['price'] = df_airbnb['price'].str.replace(',','')
df_airbnb['price'] = pd.to_numeric(df_airbnb['price'], errors='coerse')  # #error

# histogram
df_airbnb['price'].plot.hist(bins=40, title="Price Histogram")
plt.show()

# boxplot
df_airbnb['price'].plot.box(title="Price Box")
plt.show()
```

Prices are super skewed (lots of cheap rentals, few very expensive ones).
Boxplot shows many outliers at high prices.

```python
df_airbnb['price_log'] = np.log(df_airbnb['price']+1)  # +1 to avoid log(0)

df_airbnb['price_log'].plot.hist(bins=40, title="Log Price Histogram")
plt.show(). ##Log transformation compresses big numbers  histogram looks more normal.
```

```python
sns.scatterplot(data=df_airbnb, x='Beds', y='price_log')
plt.title("Price vs Beds (log)")
plt.show()

df_airbnb.groupby('Beds')['price_log'].describe()
```

1-2 beds show lots of variability.
More beds  higher mean price, but weird spikes exist.

```python
sns.scatterplot(data=df_airbnb, x='Beds', y='price_log', hue='Room Type', style='Property Type')
plt.title("Price vs Beds by Room/Property")
plt.show()

df_airbnb.groupby(['Room Type','Property Type'])['price_log'].describe()
#Entire home/apartment usually highest price.
#Other property type has highest mean & SD.
#Median seems better than mean for central tendency because of extreme outliers.
```

```python
df_rigs = pd.read_csv('./eda_assignment/data/drilling_rigs.csv', low_memory=False)
df_rigs.head()
df_rigs['time'] = pd.to_datetime(df_rigs['Month'], format='mixed')
df_rigs['time'] = pd.to_datetime(df_rigs['Month'], format='mixed')
```

Overall trend decreasing.
Some spikes & drops.

```python
df_rigs['Active Well Service Rig Count (Number of Rigs)'] = pd.to_numeric(df_rigs['Active Well Service Rig Count (Number of Rigs)'], errors='coerce')
df_rigs['diff'] = df_rigs['Active Well Service Rig Count (Number of Rigs)'].diff()

sns.lineplot(data=df_rigs, x='time', y='diff')
plt.title("Change in Rig Count Over Time")
plt.show()
```

```python
df_rigs_melt = pd.melt(df_rigs, id_vars='time',
                       value_vars=['Crude Oil and Natural Gas Rotary Rigs in Operation, Onshore (Number of Rigs)',
                                   'Crude Oil and Natural Gas Rotary Rigs in Operation, Offshore (Number of Rigs)'],
                       var_name='Rig Type', value_name='Count')

sns.lineplot(data=df_rigs_melt, x='time', y='Count', hue='Rig Type')
plt.title("Onshore vs Offshore Rigs")
plt.show()
```

Onshore usually more rigs than offshore.
Both fluctuate together over time.
